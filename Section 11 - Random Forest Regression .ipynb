# Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as  plt
import pandas as pd
```

# Importing the dataset

```python
dataset = pd.read_csv('Position_Salaries.csv')
x = dataset.iloc[:, 1:-1 ].values
y = dataset.iloc[:, -1 ].values # -1 index of last column which is the dependent variable `Profit`
```

# Training the RFR on the whole dataset

```python
from sklearn.ensemble import RandomForestClassifier
regressor = RandomForestClassifier(n_estimators = 10, random_state=0) # 10 trees & state = 0 to have the same resutls as the tutorial
regressor.fit(x, y)
```

# Predicting the new result

```python
regressor.predict([[6.5]]) #level 6.5
```

# Visualising the DTR results (for higher resolution)

```python
x_grid = np.arange(min(x), max(x), 0.1)
x_grid = x_grid.reshape((len(x_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(x_grid, regressor.predict(x_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
